Number of vissing values in each column, Filling it in with the median age, Dropping "Cabin" column because doesn't have any real valuem, Renaming "Pclass" to "Passenger_class" for more clarity, And adding "Age_Group" column that labels under eighteens as "Child" and "Adult" otherwise.

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.drop(columns=('Cabin'), inplace=True)
```

```python
df.rename(columns={'Pclass':'Passenger_class'}, inplace=True)
```

```python
def age_group_function(row):
    if row < 18:
        return ('Child')
    else:
        return ('Adult')

df['Age_Group'] = df['Age'].apply(age_group_function)
df
```

```python
df = df.dropna()
```
