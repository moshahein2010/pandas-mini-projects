## Data Cleaning
First, we will drop duplicates, fill in missing data, format dates, and convert columns to their right types.

```python
df = df[df['Quantity'] != 'ERROR']
df = df[df['Quantity'] != 'UNKNOWN']
df = df[df['Price Per Unit'] != 'ERROR']
df = df[df['Price Per Unit'] != 'UNKNOWN']
df = df[df['Total Spent'] != 'ERROR']
df = df[df['Total Spent'] != 'UNKNOWN']
df = df[df['Transaction Date'] != 'ERROR']
df = df[df['Transaction Date'] != 'UNKNOWN']
```

```python
df['Quantity'] = pd.to_numeric(df['Quantity'])
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'])
df['Total Spent'] = pd.to_numeric(df['Total Spent'])
```

I removed all data labeled as "ERROR" and "UNKNOWN" and converted the data to its correct types, such as quantity to a numeric type.

```python
df.isnull().sum()
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean()).apply(np.floor)

df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].mean()).apply(np.floor)

df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])
```

I removed all null values that matter or have real values like "Quantity", "Price Per Unit", and even "Total Spent", though it didn't have any missing values. If there is a dataframe with null values, we could get the value by multiplying "Quantity" by "Price Per Unit".

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
```

I changed the column "Transaction Date" to datetime formate.
